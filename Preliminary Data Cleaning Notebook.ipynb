```python
def convert_money(m):
    try:
        if m == 'None':
            return m
        else:
            m = m[1:]
            if m[-1] == 'K':
                m = float(m[:-1])*1000
            elif m[-1] == 'M':
                m = float(m[:-1])*1000
            return int(m)
    except:
        return float("Nan")
```

```python
df['Release Clause'] = df['Release Clause'].apply(lambda x: convert_money(x))
```
